### Target Encoding

```python
from category_encoders import TargetEncoder
```

```python
df1= pd.read_csv('Churn_Modelling_Modified.csv')
```

```python
df1.head()
```

```python
df1.drop(columns=['RowNumber','CustomerId','Surname'], axis=1, inplace=True)
```

```python
df1['Gender']= df1['Gender'].fillna('Male')
```

```python
df1.head()
```

```python
encoder = TargetEncoder()

df1['Gender Encoded']=encoder.fit_transform(df1['Gender'],df1['Exited'])
```

```python
df1.head()
```

```python
df1['Gender Encoded'].value_counts()
```

#### Benefits of TE:

- Target Encoding is a simple and quick encoding method that doesn't add to the dimensionality of the dataset. Therefore, it may be used as a first good try technique.

#### Limitations of TE:

- It is dependent on the target distribution which means target encoding requires careful validation as it can be prone to ovetfitting.
